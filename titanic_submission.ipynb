```python
import pandas as pd
```

```python
ds1 = pd.read_csv('train.csv')
```

```python
ds1.head()
```

```python
ds1.drop(['PassengerId','Name','Ticket','Cabin','Embarked','SibSp','Parch'], axis=1, inplace=True )
```

```python
ds1.head()
```

```python
ds1.isnull().sum()
```

```python
ds1['Age'].describe()
```

```python
ds1['Age'].fillna(ds1['Age'].mean(),inplace=True)
```

```python
l_sex_dummies=pd.get_dummies(ds1['Sex'],drop_first=True)
```

```python
ds1= pd.concat([ds1,l_sex_dummies],axis=1)
```

```python
ds1.head()
```

```python
ds1.drop(['Sex'], axis=1, inplace=True )
```

```python
ds1.head()
```

```python
from sklearn.preprocessing import StandardScaler
sts = StandardScaler()
```

```python
feature_scale = ['Age','Fare']
ds1[feature_scale] = sts.fit_transform(ds1[feature_scale])
```

```python
ds1.head()
```

```python
x=ds1.drop(['Survived'],axis=1)
y=ds1['Survived']
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
model_svc = SVC(C= 100,kernel='rbf')
```

```python
model_svc.fit(x,y)
```

```python
ds2 = pd.read_csv('test.csv')
```

```python
ds2.head()
```

```python
ds3=ds2.drop(['PassengerId','Name','Ticket','Cabin','Embarked','SibSp','Parch'], axis=1)
```

```python
ds3.isnull().sum()
```

```python
ds3['Age'].fillna(ds3['Age'].mean(),inplace=True)
ds3['Fare'].fillna(ds3['Fare'].mean(),inplace=True)
```

```python
l_sex_dummies=pd.get_dummies(ds3['Sex'],drop_first=True)
ds3= pd.concat([ds3,l_sex_dummies],axis=1)
ds3.drop(['Sex'], axis=1, inplace=True)
```

```python
ds3.head()
```

```python
ds3[feature_scale] = sts.fit_transform(ds3[feature_scale])
```

```python
ds3.head()
```

```python
y_predicted = model_svc.predict(ds3)
```

```python
submission = pd.DataFrame({
    "PassengerId": ds2['PassengerId'],
    "Survived": y_predicted
})
```

```python
submission.to_csv('titanic_submission_v02.csv', index=False )
```
